```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import string
import glob
from tqdm import tqdm
import random
import pickle
import string
```

```python
# Load pseudonym_mapping
with open('Pseudonym_Table/pseudonym_mapping.pkl', 'rb') as f:
    pseudonym_mapping = pickle.load(f)

# Load timeseries_mapping
with open('Pseudonym_Table/timeseries_mapping.pkl', 'rb') as f:
    timeseries_mapping = pickle.load(f)
    
#print(timeseries_mapping.items())
```

```python
# Define a random seed for reproducibility
random.seed(42)

# Load the pseudonymized DataFrame
pseudonymized_df = pd.read_csv('Pseudonym_Table/Meta_data_Table.csv', delimiter=';')

# Reverse the pseudonymization for each column
for col in pseudonymized_df.columns[1:]:
    pseudonymized_df[col] = pseudonymized_df[col].map({v: k for k, v in pseudonym_mapping.items()})

# Add timeseries_mapping as an additional column
pseudonymized_df['timeseriesname'] = pseudonymized_df['Dataset'].map(timeseries_mapping)
```
